# DETERMINING THE RELATION BETWEEN DRUG CONSUMPTION AND TERRORISM

```python
dt=pd.read_excel("GTD/globalterrorismdb_0718dist.xlsx")
```

```python
dt1=dt.groupby(['country_txt'] ,sort= True).size().to_frame('frequency')
#dt1=dt['country_txt'].value_counts().to_dict()
dt1=dt1.sort_values(by='frequency', ascending=False)
dt1=dt1.reset_index()
dt1=dt1.rename(columns ={'country_txt': 'Country'})
```

```python
dt1
```

```python
dg=pd.read_excel("drug_data/1.2_Prevalence_of_drug_use_in_the_general_population_-_national_data.xls",header=2)
```

```python
dg=dg.groupby('Country').max().reset_index()
dg=dg.drop(['Low','High','Male','Female','Year'],axis=1)
```

```python
dg1=pd.read_excel("drug_data/1.2cocaine.xlsx",header=2)
dg2=pd.read_excel("drug_data/1.2amphetamines.xlsx",header=2)
dg3=pd.read_excel("drug_data/1.2transq.xlsx",header=2)
dg4=pd.read_excel("drug_data/1.2opiate.xlsx",header=2)
dg1=dg1.groupby('Country').max().reset_index()
dg1=dg1.drop(['Low','High','Male','Female','Year','Age','Source'],axis=1)
dg2=dg2.groupby('Country').max().reset_index()
dg2=dg2.drop(['Low','High','Male','Female','Year','Age','Source'],axis=1)
dg3=dg3.groupby('Country').max().reset_index()
dg3=dg3.drop(['Low','High','Male','Female','Year','Age','Source'],axis=1)
dg4=dg4.groupby('Country').max().reset_index()
dg4=dg4.drop(['Low','High','Male','Female','Year','Age','Source'],axis=1)
```

```python
df = pd.concat([dg, dg1,dg2,dg3,dg4]).groupby('Country').max()
df=df.reset_index()
```

```python
df
```

```python
final=pd.merge(df,dt1,how='inner',on=['Country'])
```
