# PREPROCESSING POPULATION TO BRING RESOLUTION TO 0.1

```python
df=pd.read_csv('data\Topography\SRTM_RAMP2_TOPO_2000-02-11_rgb_3600x1800.SS.CSV',index_col=0)
```

```python
df.head()
```

```python
a=[str(round(float(x),2))[:-1] for x in list(df)]
```

```python
df.columns=a
df.head()
```

```python
df.at[89.95,'-179.9']
```

```python
df=df.reset_index()
```

```python
print(df['lat/lon'])
```

```python
ind=[ float(str(round(float(x),2))[:-1]) for x in df['lat/lon']]
```

```python
n = df.columns[0]
df.drop(n, axis = 1, inplace = True)
df.insert(loc=0, column='lat/lon', value=ind)
```

```python
df
```

```python
df=df.set_index('lat/lon')
```

```python
df.head()
```

```python
df.at[89.9,'-179.9']
```

```python
df.to_csv(r'data\Topography\ResampledTopography.csv')
```

```python
data=[]

for i in list(df.index):
    for j in list(df.columns):
        data.append([ i , j , df.at[i,j] ])
```

```python
top = pd.DataFrame(data,columns=['latitude','longitude','topography'])
```

```python
top['longitude']=top['longitude'].astype('float32')
```

```python
top.to_csv('data\Topography\TopographyData.csv',index=False)
```
