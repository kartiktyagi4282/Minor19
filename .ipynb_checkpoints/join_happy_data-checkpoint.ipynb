```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('data/World Happiness Report/happy.csv')
df.shape
```

```python
df.head(10)
```

```python
inp=pd.read_csv('data/gtd_geoecon_pop_geoepr_topography_nightlight_join.csv')
```

```python
inp.shape
```

```python
inp.head(10)
```

```python
df[(df['latitude']== 18.5) & (df['longitude']== -70.0)]
```

```python
l=[]

for i in range(inp.shape[0]):
    x=inp.loc[i].latitude
    y=inp.loc[i].longitude
    if i%10000==0:
        print(i)
        
    try:
        a=df[(df['latitude']== x) & (df['longitude']== y)]
        
        if a.empty==True:
            l.append(99999)
        else:
            ind=a.index[0]
            l.append(df.at[ind,'happy'])    
    except KeyError:
        l.append(99999)
        pass
```

```python
len(l)
```

```python
inp['happy_index']=l
```

```python
inp.drop(columns=['Unnamed: 0'],inplace=True)
```
