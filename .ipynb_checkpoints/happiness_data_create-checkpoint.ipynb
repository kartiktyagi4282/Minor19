```python
import pandas as pd
import numpy as np
```

```python
dfw=pd.read_csv('data/World Happiness Report/2015.csv')
dfw.shape
```

```python
gtd=pd.read_excel('data/GTD/globalterrorismdb_0718dist.xlsx')
gtd.shape
```

```python
gtd.columns
```

```python
dfw.head(10)
```

```python
df=gtd[['country_txt','latitude','longitude']]
```

```python
df.head(10)
```

```python
df['latitude']=df['latitude'].round(1)
df['longitude']=df['longitude'].round(1)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.dropna(inplace=True)
df.reset_index(drop=True,inplace=True)
df.info()
```

```python
dfw.info()
```

```python
l=[]
cnt=0
for i in range(df.shape[0]):
    con=df.loc[i].country_txt
    
    if i%10000==0:
        print(i)
        
    try:
        a=dfw[dfw['Country']== con]
        if a.empty==True:
            #print(a)
            l.append(99999)
            print(con)
            cnt+=1
        else:
            ind=a.index[0]
            l.append(a.at[ind,'Happiness Score'])
    except KeyError:
        l.append(99999)
        pass
print(cnt)
```

```python
df['happy']=l
```

```python
df.head()
```

```python
df.to_csv('data/World Happiness Report/happy.csv',index=False)
```
