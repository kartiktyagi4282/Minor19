# Applying group by on the "gtd_gecon_pop_geoEPR_join(temp).csv" file to generate a file "gtd_geoepr_join(with_counts).csv"

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r'data\gtd_gecon_pop_geoEPR_join(temp).csv')
```

```python
df.drop('field_1',axis=1,inplace=True)
```

```python
df.head()
```

```python
df1=df.groupby(['latitude','longitude']).gwgroupid.nunique()
df1
```

```python
df2=df1.reset_index()
df2
```

```python
df2.to_csv('data/gtd_geoepr_join(with_counts).csv',index=False)
```
